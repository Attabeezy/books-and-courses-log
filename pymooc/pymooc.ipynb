```python
# the next leap year

year = int(input("Year: "))
init_year = year
year += 1

while True:
    if year % 400 == 0:
        print(f"The next leap year after {init_year} is {year}")
        break
    if year % 100 == 0:
        year += 1
        continue
    if year % 4 == 0:
        print(f"The next leap year after {init_year} is {year}")
        break
    year += 1
```

```python
numbers = []

def positive_negative(numbers):
    pos = []
    neg = []
    for number in numbers:
        if number >= 0:
            pos.append(number)
        else:
            neg.append(number)
    return pos, neg

print("Please type in integer numbers. Type in 0 to finish.")

while True:
    num = int(input("Number: "))
    if num == 0:
        print("... the program asks for numbers")
        print(f"Numbers typed in {len(numbers)}")
        print(f"The sum of the numbers is {sum(numbers)}")
        print(f"The mean of the numbers is {sum(numbers)/len(numbers)}")
        pos, neg = positive_negative(numbers)
        print(f"Positive numbers {len(pos)}")
        print(f"Negative numbers {len(neg)}")
        break
    else:
        numbers.append(num)
        
        continue
```

```python
def squared(string, size):
    square = size ** 2
    len_string = len(string)
    
    
    if len_string % square == 0:
        new_string = string * square
    else:
        new_string = (string * square)[:square]
    for i in range(0, len(new_string), size):
        print(new_string[i:i+size])

if __name__ == "__main__":
    squared("attabeezy", 8)
```
